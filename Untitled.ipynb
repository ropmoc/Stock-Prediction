```python
def mean_absolute_percentage_error(y_true, y_pred): 
    y_true, y_pred = numpy.array(y_true), numpy.array(y_pred)
    return numpy.mean(numpy.abs((y_true - y_pred) / y_true)) * 100
```

```python
split_size = round(len(jopen)*0.6)

com_input= [7] 

X_ans=[]
Y_ans=[]
masukan=[]
keluaran=[]

for i in range(len(com_input)):
    for j in range(len(jopen)-com_input[i]): 
        X=list(jopen)[j:j+com_input[i]]
        Y=list(jopen)[j+com_input[i]]
        X_ans.append(X)
        Y_ans.append(Y)
    mas_temp=array(X_ans)
    kel_temp=array(Y_ans)
    masukan.append(mas_temp)
    keluaran.append(kel_temp)
    X_ans.clear()
    Y_ans.clear()

# split train dan test

train_in=[]
train_out=[]
test_in=[]
test_out=[]
counter=0
for i in range(len(com_input)):
    size=455
    s = size - split_size - com_input[i]
    train_in.append(masukan[i][:s])
    train_out.append(keluaran[i][:s])
    test_in.append(masukan[i][s:])
    test_out.append(keluaran[i][s:])
    size = size-1
    split_size = split_size-com_input[i] 
    counter=counter+1
```

```python
model = Sequential()
model.add(Dense(100, activation='selu', input_dim=7))
model.add(Dense(1, activation='relu'))
model.compile(optimizer='adam', loss='mape', metrics=['mape'])
history = model.fit(train_in[0], train_out[0], epochs=100, verbose=0)
mape = model.evaluate(test_in[0],test_out[0])
probabilities = (model.predict(test_in[0], verbose=0))
predictions = [float(x.round()) for x in probabilities]
test_mapelist = mean_absolute_percentage_error(test_out[0],predictions) 
print(test_mapelist)
```
